```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data=pd.get_dummies(data)
```

```python
train=data[0:7999]
```

```python
test=data[8000:]
```

```python
test.head()
```

```python
 x_train=train.drop('Item_Outlet_Sales',axis=1)
    
```

```python
y_train=train['Item_Outlet_Sales']
```

```python
x_test=test.drop('Item_Outlet_Sales',axis=1)
```

```python
truep_test=test['Item_Outlet_Sales']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg=LinearRegression()
```

```python
x_train=pd.get_dummies(x_train)
```

```python
x_train
```

```python
x_train.fillna(0,inplace=True)
```

```python
x_test.fillna(0,inplace=True)
```

```python
lreg.fit(x_train,y_train)
```

```python
pred=lreg.predict(x_test)
```

```python
lreg.score(x_test,truep_test)
```

```python
lreg.score(x_train,y_train)
```

```python
rmse_test=np.sqrt(np.mean(np.power((np.array(truep_test)-np.array(pred)),2)))
```

```python
rmse_train=np.sqrt(np.mean(np.power((np.array(y_train)-np.array(lreg.predict(x_train))),2)))
```

```python
print(rmse_test)
```

```python
print(rmse_train)
```
